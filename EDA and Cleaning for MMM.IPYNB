# Data Cleaning

```python
# read the csv data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

products = pd.read_csv('product_table_supp.csv',encoding = "ISO-8859-1")
transactions = pd.read_csv('transaction_table_supp.csv',encoding = "ISO-8859-1")
```

```python
pd.options.display.max_columns = None # this allows us to see all columns for head
products.head()
```

```python
pd.options.display.max_columns = None # this allows us to see all columns for head
transactions.head()
```

```python
transactions.sort_values(by = 'tran_dt')
```

## Get tran_wk and calculate prices

```python
import datetime as dt
transactions['tran_dt']= pd.to_datetime(transactions['tran_dt'])
#transactions['tran_wk']= transactions['tran_dt'].dt.to_period('W').apply(lambda r: r.end_time)
```

```python
transactions.sort_values(by = 'tran_dt')
```

```python
transactions['tran_wk']= transactions['tran_wk'].astype(str).str.slice(start=0,stop=10)
```

```python
transactions.head()
```

```python
# Calculate price
transactions["pocket_price"]=round(transactions["tran_prod_paid_amt"]/transactions["tran_prod_sale_qty"],2)
transactions["list_price"]=round(transactions["prod_unit_price"],2)
transactions["discount"]=transactions["list_price"]-transactions["pocket_price"]
transactions.tail()
```

## Get weekly transactions

```python
transactions_wk=transactions.groupby(['tran_wk','prod_id']).agg({'tran_prod_sale_qty':'sum',
                                                                 'tran_prod_paid_amt':'sum',
                                                                 'pocket_price':'mean',
                                                                'list_price':'mean',
                                                                'discount':'mean'}).reset_index().sort_values(by='tran_wk',ascending=True)
```

```python
transactions_wk.head()
```

```python
transactions_wk['pocket_price']=round(transactions_wk['pocket_price'],2)
transactions_wk['list_price']=round(transactions_wk['list_price'],2)
transactions_wk['discount']=transactions_wk['list_price']-transactions_wk['pocket_price']
```

```python
transactions_wk.columns=['tran_wk', 'prod_id', 'sale_qty', 'sale_amt',
       'pocket_price', 'list_price', 'discount']
```

## Get all csv info

### Get info of holiday & seasonality

```python
# Read holiday, seasonality
holiday=pd.read_csv('holiday.csv')
seasonality=pd.read_csv('seasonality.csv')
```

```python
holiday
```

```python
complete_df = pd.merge(transactions_wk,holiday,left_on = 'tran_wk', right_on = 'tran_wk', how = 'left')
complete_df = pd.merge(complete_df,seasonality,left_on = 'tran_wk', right_on = 'tran_wk', how = 'left')
```

```python
## Holiday change NaN into 0, others into 1
complete_df['holiday'].fillna('No holiday',inplace=True)
#complete_df['holiday'] = np.where(complete_df['holiday'] !=0, 1, complete_df['holiday'])
```

```python
complete_df
```

### Get info of flyer & store display

```python
promo_ad=pd.read_csv('promo_ad.csv')
```

```python
flyer=promo_ad[promo_ad['vehicle']=='Flyer'].drop(["vehicle","unit"],axis=1)
flyer.columns=['tran_wk','flyer','prod_id']
flyer['prod_id']=flyer['prod_id'].astype('int')
flyer.head()
```

```python
complete_df = pd.merge(complete_df,flyer,left_on = ['tran_wk','prod_id'], right_on = ['tran_wk','prod_id'], how = 'left')
```

```python
## Flyer change NaN into 0
complete_df['flyer'].fillna(0,inplace=True)
complete_df.head()
```

```python
storedisplay=promo_ad[promo_ad['vehicle']=='Store Display'].drop(["vehicle","unit"],axis=1)
storedisplay.columns=['tran_wk','store_display','prod_id']
storedisplay['prod_id']=storedisplay['prod_id'].astype('int')
storedisplay.head()
```

```python
complete_df = pd.merge(complete_df,storedisplay,left_on = ['tran_wk','prod_id'], right_on = ['tran_wk','prod_id'], how = 'left')
```

```python
## Store display change NaN into 0
complete_df['store_display'].fillna(0,inplace=True)
```

## Get the impressions & reach for paid search & web display

```python
# take the info of paid search
paid_search=promo_ad[promo_ad['vehicle']=='Paid Search'].drop(["vehicle","unit",'prod_assoc'],axis=1)
paid_search.columns=['tran_wk','paid_search_impressions']
paid_search.head()
```

```python
complete_df = pd.merge(complete_df,paid_search,left_on = ['tran_wk'], right_on = ['tran_wk'], how = 'left')
```

```python
complete_df = pd.merge(complete_df,web_display,left_on = ['tran_wk'], right_on = ['tran_wk'], how = 'left')
```

```python
## web_display change NaN into 0
complete_df['web_display_impressions'].fillna(0,inplace=True)
complete_df['web_display_impressions'] = complete_df['web_display_impressions']/3
```

Add GRP Values for TV and Radio Advertisements

```python
Radio=promo_ad[promo_ad['vehicle']=='Radio'].drop(["vehicle","unit",'prod_assoc'],axis=1)
Radio.columns=['tran_wk','Radio GRP']
Radio.head()
```
